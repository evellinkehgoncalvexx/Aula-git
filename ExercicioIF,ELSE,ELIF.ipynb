```python
salario = int(input("Digite seu salário: R$ "))

if 1500 <= salario <= 2500:
  aumento = salario * 0.25
  novo_salario = salario + aumento
  print(f"25% do seu salário é R${aumento}")
  print(f"Seu novo salário é R${novo_salario}")

elif 2600 <= salario <= 4900:
  aumento = salario * 0.15
  novo_salario = salario + aumento
  print(f"15% do seu salário é R${aumento}")
  print(f"Seu novo salário é R${novo_salario}")

else:
  salario > 5000
  print("Sem aumento")
  print(f"Salário permanece R${salario}")

```
