```python
#Exercício 4

prova1= float(input("Digite a nota da primeira prova: "))
prova1 = prova1 * 2
prova2= float(input("Digite a nota da segunda prova: "))
prova2 = prova2 * 3
prova3= float(input("Digite a nota da terceira prova: "))
prova3 = prova3 * 5

media= (prova1 + prova2 + prova3) /10
print ("Seu resultado é: ", media)
```

```python
#Exercício 5
altura= float(input("Insira a altura da caixa d'água: "))
largura= float(input("Insira a largura da caixa d'água: "))
profundidade= float(input("Insira a profundidade da caixa d'água: "))

volume= altura * largura * profundidade
print ("O volume da caixa d'água é: ", volume)
```
