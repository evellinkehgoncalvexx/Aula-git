```python
# Exercício 1

num1= int(input("Digite o primeiro número: "))
num2= int(input("Digite o segundo número: "))

if num1 > num2:
    print("O maior número é: ", num1)
else:
    print("O maior número é: ", num2)
```

```python
# Exercício 2

nome1= str(input("Digite o seu nome: "))
altura1= float(input("Digite a sua altura: "))

nome2= str(input("Digite o seu nome: "))
altura2= float(input("Digite a sua altura: "))

if altura1 > altura2:
    print(f"A pessoa mais alta é: {nome1}, com {altura1} metros de altura.")
else:
    print(f"A pessoa mais alta é: {nome2}, com {altura2} metros de altura.")
```

```python
# Exercício 3

compra= float(input("Digite o valor da compra: "))
desconto= 10 * compra / 100
novoValor = compra - desconto

if compra >= 100:
  print("O total a ser pago pelo cliente é: ", novoValor)

else:
  print("O total a ser pago pelo cliente é: ", compra)
```
